```python
import yfinance as yf
from matplotlib import pyplot as plt

# Load the MSFT Stock closing prices data from Yahoo Finance for the period 2018-01-01 to Latest date
msft = yf.Ticker("MSFT")
hist = msft.history(period="max")
hist = hist[hist.index >= '2018-01-01']
hist = hist[['Close']]
hist = hist.rename(columns={'Close': 'MSFT'})
hist = hist.dropna()
hist
```

```python
# Save the data to a CSV file
hist.to_csv('MSFT.csv')
```

```python
# Plot the data
hist.plot(figsize=(10, 5))
plt.title('MSFT Stock Price')
plt.ylabel('Price')
plt.grid()
plt.savefig('MSFT.png')
plt.show()
```
